```python
import numpy as np
import pandas as pd
%matplotlib inline  
import matplotlib
import matplotlib.pyplot as plt  
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from collections import Counter
```

### Van Diagram

```python
for_="Organization"
ff="organization"
ext="org"

dec_org_d = pd.read_csv('../outputs/{}/venn/venn_decorators_{}.csv'.format(ff,ext))
dec_org = dec_org_d[[for_]]
func_org_d = pd.read_csv('../outputs/{}/venn/venn_functional_{}.csv'.format(ff,ext))
func_org = func_org_d[[for_]]
dyn_org_d = pd.read_csv('../outputs/{}/venn/venn_dynamic_{}.csv'.format(ff,ext))
dyn_org = dyn_org_d[[for_]]
```

```python
#Get the set of the dataframe
def gset(a,b):
    return set(a[b])
dyn_set = gset(dyn_org,for_)
func_set = gset(func_org,for_)
dec_set = gset(dec_org,for_)
```

```python
subsets = (1, 1, 0.2, 1, 0.2, 0.2, 0.1)

colors = ['darkviolet','deepskyblue','blue']
v2 = venn3(subsets=subsets, alpha=0.7,set_colors=colors)
labels = ['Dynamic', 'Functional', 'Decorators']
labels = ['100', '101', '110', '010', '001', '011', '111']
for label in labels:
    v2.get_label_by_id(label).set_text(label)
   
```

```python

plt.figure(figsize=(10,10))
ax = plt.gca()

colors = ['darkviolet','deepskyblue','blue']
labels=('','','')
sets=[dyn_set, func_set, dec_set]
v = venn3(subsets=sets, set_labels=labels, alpha=0.7,set_colors=colors)
i=0
for text in v.set_labels:
  text.set_color(colors[i])
  i+=1
    
id_labels = ['100', '101', '110', '010', '001', '011', '111']
subset_labels=[v.get_label_by_id('110'),v.get_label_by_id('111'),v.get_label_by_id('011'),v.get_label_by_id('010')]
for i in id_labels:
  text = v.get_label_by_id(i)
  text.set_color('white')
  text.set_fontsize(16)
  text.set_fontweight('bold')
set_ids=id_labels
h = [] 
for i in set_ids:
    h.append(v.get_patch_by_id(i)) 
l = ['Dyn Only','Dyn & Dec','Func & Dyn','Func Only','Dec Only','Func & Dec','Func & Dyn & Dec']
ax.legend(handles=h, labels=l, title="Legend",loc='lower left') 

plt.title("{} Feature Count".format(for_))
i = 0

plt.gca().set_facecolor('azure')
plt.gca().set_label('all files')

plt.gca().set_axis_on()
plt.show()
```

### Histogram

```python


func_freq_d = pd.read_csv('../outputs/functional_frequency_rank.csv')
func_freq = func_freq_d[['freq']]
func_freq_name = func_freq_d[['funcname']]
```

```python
def valuelabel(func_freq,func_freq_name):
    for i in range(len(func_freq)):
        plt.text(i,func_freq_name[i],func_freq_name[i], ha = 'center')
```

```python
plt.figure(figsize=(10,10))
y_pos = np.arange(len(func_freq_name))
plt.bar(y_pos, func_freq['freq'], color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.xticks(y_pos, func_freq_name['funcname'])

plt.title('Functional Feature Ranking')

valuelabel(func_freq_name['funcname'], func_freq['freq'])    
plt.gca().set_facecolor('azure')
plt.show()
```

```python
func_freq_d = pd.read_csv('../outputs/decorators_frequency_rank.csv')
func_freq = func_freq_d[['freq']]
func_freq_name = func_freq_d[['funcname']]

plt.figure(figsize=(20,10))
y_pos = np.arange(len(func_freq_name))
plt.bar(y_pos, func_freq['freq'], color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.xticks(y_pos, func_freq_name['funcname'], rotation=90)

plt.title('Decorators Feature Ranking')

valuelabel(func_freq_name['funcname'], func_freq['freq'])    
plt.gca().set_facecolor('azure')
plt.show()
```

```python
func_freq_d = pd.read_csv('../outputs/dynamic_frequency_rank.csv')
func_freq = func_freq_d[['freq']]
func_freq_name = func_freq_d[['funcname']]

plt.figure(figsize=(10,10))
y_pos = np.arange(len(func_freq_name))
plt.bar(y_pos, func_freq['freq'], color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
plt.xticks(y_pos, func_freq_name['funcname'])

plt.title('Dynamic Feature Ranking')

valuelabel(func_freq_name['funcname'], func_freq['freq'])    
plt.gca().set_facecolor('azure')
plt.show()
```
