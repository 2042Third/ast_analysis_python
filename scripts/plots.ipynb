```python
import numpy as np
import pandas as pd
%matplotlib inline  
import matplotlib
import matplotlib.pyplot as plt  
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
```

### Van Diagram

```python
dec_org_d = pd.read_csv('../outputs/organization/venn/venn_decorators_org.csv')
dec_org = dec_org_d[['Organization']]
func_org_d = pd.read_csv('../outputs/organization/venn/venn_functional_org.csv')
func_org = func_org_d[['Organization']]
dyn_org_d = pd.read_csv('../outputs/organization/venn/venn_dynamic_org.csv')
dyn_org = dyn_org_d[['Organization']]
```

```python
#Get the set of the dataframe
def gset(a,b):
    return set(a[b])
dyn_set = gset(dyn_org,"Organization")
func_set = gset(func_org,"Organization")
dec_set = gset(dec_org,"Organization")
```

```python
subsets = (1, 1, 0.2, 1, 0.2, 0.2, 0.1)
v2 = venn3(subsets=subsets)
labels = ['Dynamic', 'Functional', 'Decorators']
labels = ['100', '101', '110', '010', '001', '011', '111']
for label in labels:
    v2.get_label_by_id(label).set_text(label)
   
```

```python

# print(dyn_set)
v = venn3([dyn_set, func_set, dec_set], ('Dynamic', 'Functional', 'Decorators'))
# fig = plt.figure()
# v = venn3([dyn_set, Functional, dec_set], ('Dynamic',  'Decorators'))

plt.annotate('Func Dyn', xy=v.get_label_by_id('110').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))
plt.annotate('Func Dyn Dec', xy=v.get_label_by_id('111').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))
plt.annotate('Func Dec', xy=v.get_label_by_id('011').get_position() - np.array([0, 0.05]), xytext=(-70,-70),
ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='gray'))

plt.title("Organization Feature Frequency")

plt.show()
```
